```python
import yfinance as yf
import pandas as pd
import numpy as np
nvda = yf.Ticker("NVDA")

# get historical market data
hist = nvda.history(period="6mo")
hist.shape
```

```python
hist
```

```python
hist['var']=hist['Close']-hist['Open']
hist['p_var']=(hist['var']/hist['Close'])*100
hist['f_alza'] = hist['var'].apply(lambda x: 1 if x>0 else 0)
```

```python
hist
```

```python
hist['f_alza'].mean()
```

```python
hist['p_var'].describe()
```

```python
tickers = ['NVDA']
for ticker in tickers:
    ticker_yahoo = yf.Ticker(ticker)
    data = ticker_yahoo.history()
    last_quote = data['Close'].iloc[-1]
    print(ticker, last_quote)
```
